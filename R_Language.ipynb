In R, you can assign values to variables using the `<-` operator (or `=`). Let's create a few variables and print their values.

In R, a **vector** is a sequence of data elements of the same basic type. Think of it as a list of numbers, characters, or logical values. You can create vectors using the `c()` function, which stands for 'combine' or 'concatenate'.

```r
# Create a numeric vector
numeric_vector <- c(1, 5, 9, 13, 17)

# Create a character vector
character_vector <- c("apple", "banana", "cherry")

# Create a logical vector
logical_vector <- c(TRUE, FALSE, TRUE, FALSE)

# Print the vectors
print(numeric_vector)
print(character_vector)
print(logical_vector)
```

Vectors are very versatile, and you can perform various operations on them, such as arithmetic calculations on numeric vectors or accessing specific elements.

A **Data Frame** in R is a list of vectors of equal length. Each vector serves as a column, and each column can have a different data type. This is similar to a table in a database or a spreadsheet.

```r
# Create a Data Frame
data_frame_example <- data.frame(
  Name = c("Alice", "Bob", "Charlie", "Diana"),
  Age = c(24, 30, 28, 22),
  IsStudent = c(TRUE, FALSE, FALSE, TRUE)
)

# Print the Data Frame
print(data_frame_example)

# Get the structure of the Data Frame
print(str(data_frame_example))

# Access a column using the $ operator or double square brackets [[]]
print("\nNames:")
print(data_frame_example$Name)

print("\nAges:")
print(data_frame_example[["Age"]])

# Access rows
print("\nFirst row:")
print(data_frame_example[1, ])

print("\nRows 2 and 3:")
print(data_frame_example[2:3, ])
```

You can perform standard arithmetic operations directly on numeric vectors. R applies these operations element-wise.

```r
# Perform arithmetic operations on numeric_vector
print("Original numeric_vector:")
print(numeric_vector)

print("\nAdd 5 to each element:")
print(numeric_vector + 5)

print("\nMultiply each element by 2:")
print(numeric_vector * 2)

print("\nSubtract a vector from another (element-wise):")
vector_b <- c(1, 2, 3, 4, 5)
print(numeric_vector - vector_b)
```

You can also access individual elements or subsets of a vector using square brackets `[]`.

```r
# Access elements of numeric_vector
print("Original numeric_vector:")
print(numeric_vector)

print("\nFirst element:")
print(numeric_vector[1])

print("\nElements at index 2 and 4:")
print(numeric_vector[c(2, 4)])

print("\nElements from index 3 to 5:")
print(numeric_vector[3:5])

print("\nAll elements except the first:")
print(numeric_vector[-1])

print("\nElements greater than 10:")
print(numeric_vector[numeric_vector > 10])
```

In R, a **function** is a block of organized, reusable code that is used to perform a single, related action. Functions provide better modularity for your application and a high degree of code reusing.

You define a function using the `function()` keyword, followed by the function's arguments in parentheses, and then the body of the function within curly braces `{}`. The `return()` statement is used to send a result back from the function, though it's often optional for simple functions that return the last evaluated expression.

```r
# Define a simple function that adds two numbers
add_numbers <- function(a, b) {
  sum_result <- a + b
  return(sum_result)
}

# Call the function with some values
result1 <- add_numbers(10, 5)
print(paste("The sum of 10 and 5 is:", result1))

# Define a function with a default argument
greet <- function(name = "World") {
  message <- paste("Hello,", name, "!")
  print(message)
}

# Call the greet function
greet()
greet("Alice")

# Define a function to calculate the square of a number
square <- function(x) {
  return(x^2)
}

# Call the square function
num <- 7
squared_num <- square(num)
print(paste("The square of", num, "is:", squared_num))
```

Conditional statements in R allow you to control the flow of your program based on certain conditions. The most common conditional statement is the `if` statement, which can be extended with `else if` and `else` clauses.

```r
# Basic if statement
x <- 10

if (x > 5) {
  print("x is greater than 5")
}

# if-else statement
y <- 3

if (y %% 2 == 0) {
  print("y is an even number")
} else {
  print("y is an odd number")
}

# if-else if-else statement
z <- -5

if (z > 0) {
  print("z is positive")
} else if (z < 0) {
  print("z is negative")
} else {
  print("z is zero")
}
```

Loops are fundamental programming constructs that allow you to execute a block of code multiple times. R provides several types of loops, including `for` loops and `while` loops.

### `for` Loop
A `for` loop iterates over a sequence (like a vector or list) or a specified number of times. It's ideal when you know in advance how many times you want to loop.

### `while` Loop
A `while` loop repeatedly executes a block of code as long as a specified condition is true. It's useful when you don't know the exact number of iterations beforehand, but rather want to loop until a certain condition is met.

```r
# Example of a for loop
print("For loop example:")
for (i in 1:5) {
  print(paste("Current iteration (for loop):", i))
}

# Example of iterating over a vector with a for loop
print("\nFor loop over a vector:")
fruits <- c("apple", "banana", "cherry")
for (fruit in fruits) {
  print(paste("I like", fruit))
}

# Example of a while loop
print("\nWhile loop example:")
j <- 1
while (j <= 3) {
  print(paste("Current iteration (while loop):", j))
  j <- j + 1
}
```

In R, a **List** is a generic object that contains an ordered collection of objects. Unlike atomic vectors, which can only hold elements of a single data type, lists can combine elements of different data types and structures. This makes them very versatile for storing heterogeneous data.

```r
# Create a simple list with different data types
my_list <- list(
  name = "John Doe",
  age = 30,
  is_student = FALSE,
  grades = c(95, 88, 92, 79),
  address = list(street = "123 Main St", city = "Anytown"),
  data_frame_part = data.frame(ID = c(1,2), Value = c("A","B"))
)

# Print the entire list
print("Full List:")
print(my_list)

# Get the structure of the list
print("\nStructure of the List:")
print(str(my_list))

# Access elements by name
print("\nAccessing elements by name:")
print(my_list$name)
print(my_list[['grades']])

# Access elements by position
print("\nAccessing elements by position:")
print(my_list[[2]]) # The second element (age)

# Access elements within nested structures
print("\nAccessing nested elements:")
print(my_list$address$city)
print(my_list$data_frame_part$Value)

# Modify an element
my_list$age <- 31
print("\nUpdated age:")
print(my_list$age)
```

In R, a **Matrix** is a two-dimensional rectangular dataset. It's essentially a collection of elements of the same data type (numeric, character, or logical) arranged in rows and columns. Matrices are widely used in linear algebra and statistical computations.

```r
# Create a simple 3x3 numeric matrix
matrix_a <- matrix(c(1, 2, 3, 4, 5, 6, 7, 8, 9), nrow = 3, ncol = 3, byrow = TRUE)
print("Matrix A (by row):")
print(matrix_a)

# Create another matrix, filling by column (default)
matrix_b <- matrix(c(10, 11, 12, 13, 14, 15), nrow = 2, ncol = 3)
print("\nMatrix B (by column):")
print(matrix_b)

# Get dimensions of a matrix
print("\nDimensions of Matrix A:")
print(dim(matrix_a))

# Access elements of a matrix
print("\nElement at row 2, column 3 of Matrix A:")
print(matrix_a[2, 3])

print("\nFirst row of Matrix A:")
print(matrix_a[1, ])

print("\nThird column of Matrix A:")
print(matrix_a[, 3])

# Matrix operations (element-wise)
print("\nMatrix A + 10:")
print(matrix_a + 10)

# Matrix multiplication (linear algebra)
matrix_c <- matrix(c(1, 0, 0, 1), nrow = 2, ncol = 2, byrow = TRUE)
matrix_d <- matrix(c(5, 6, 7, 8), nrow = 2, ncol = 2, byrow = TRUE)
print("\nMatrix C:")
print(matrix_c)
print("\nMatrix D:")
print(matrix_d)
print("\nMatrix C %*% Matrix D (multiplication):")
print(matrix_c %*% matrix_d)
```

In R, a **Factor** is a data structure used to store categorical data. They are built on top of integer vectors, where each integer represents a specific category or 'level'. Factors are particularly useful for statistical modeling and when you want to treat data as discrete categories rather than continuous numerical values.

```r
# Create a character vector representing blood types
blood_type_vector <- c("A", "B", "AB", "O", "A", "O", "B")
print("Original character vector:")
print(blood_type_vector)

# Convert the character vector to a factor
blood_type_factor <- factor(blood_type_vector)
print("\nConverted to factor:")
print(blood_type_factor)

# Check the levels of the factor
print("\nLevels of the factor:")
print(levels(blood_type_factor))

# Create an ordered factor (e.g., for education levels)
education_vector <- c("High School", "College", "Graduate", "High School", "College")
education_factor <- factor(education_vector,
                          levels = c("High School", "College", "Graduate"),
                          ordered = TRUE)
print("\nOrdered factor (Education Levels):")
print(education_factor)
print("\nLevels of the ordered factor:")
print(levels(education_factor))

# Performing operations (like counting occurrences)
print("\nCount of each blood type:")
print(table(blood_type_factor))
```

### Data Manipulation with `dplyr`

`dplyr` provides a consistent set of verbs that help you solve the most common data manipulation challenges. If you don't have it installed, you'll need to run `install.packages("dplyr")` first. Then, we load it using `library(dplyr)`.

```r
# Install dplyr if not already installed (uncomment and run if needed)
# install.packages("dplyr")

# Load the dplyr package
library(dplyr)

# We'll use our existing data_frame_example
print("Original Data Frame:")
print(data_frame_example)

# --- Common dplyr verbs ---

# 1. filter(): Select rows based on conditions
print("\nFiltered: Students who are older than 25:")
filtered_df <- data_frame_example %>%
  filter(IsStudent == TRUE, Age > 25)
print(filtered_df)

# 2. select(): Choose columns by name
print("\nSelected: Only Name and Age columns:")
selected_df <- data_frame_example %>%
  select(Name, Age)
print(selected_df)

# 3. mutate(): Add new columns or modify existing ones
print("\nMutated: Adding a 'BirthYear' column (approximate):")
mutated_df <- data_frame_example %>%
  mutate(BirthYear = 2024 - Age)
print(mutated_df)

# 4. arrange(): Order rows by a column
print("\nArranged: Ordered by Age (descending):")
arranged_df <- data_frame_example %>%
  arrange(desc(Age))
print(arranged_df)

# 5. group_by() and summarize(): Group data and calculate summary statistics
# Let's create a slightly larger dataframe for a better example
some_data <- data.frame(
  Category = c("A", "B", "A", "B", "A"),
  Value = c(10, 15, 12, 18, 11)
)

print("\nOriginal 'some_data' for grouping:")
print(some_data)

print("\nGrouped and Summarized: Mean value per category:")
summarized_data <- some_data %>%
  group_by(Category) %>%
  summarize(MeanValue = mean(Value), Count = n())
print(summarized_data)
```

### Basic Statistical Analysis
R provides a comprehensive set of functions for performing basic statistical analysis, which are crucial for understanding your data's characteristics. Here are some of the most commonly used functions:

*   `mean()`: Calculates the arithmetic mean.
*   `median()`: Calculates the median.
*   `sd()`: Calculates the standard deviation.
*   `var()`: Calculates the variance.
*   `min()`: Finds the minimum value.
*   `max()`: Finds the maximum value.
*   `sum()`: Calculates the sum of all values.
*   `summary()`: Provides a comprehensive summary of a numeric vector, including min, max, mean, median, and quartiles.
*   `cor()`: Calculates the correlation between two numeric vectors.

```r
# We'll use the 'numeric_vector' created earlier: c(1, 5, 9, 13, 17)
print("Original numeric_vector:")
print(numeric_vector)

# Mean
mean_val <- mean(numeric_vector)
print(paste("Mean:", mean_val))

# Median
median_val <- median(numeric_vector)
print(paste("Median:", median_val))

# Standard Deviation
sd_val <- sd(numeric_vector)
print(paste("Standard Deviation:", sd_val))

# Variance
var_val <- var(numeric_vector)
print(paste("Variance:", var_val))

# Minimum and Maximum
min_val <- min(numeric_vector)
max_val <- max(numeric_vector)
print(paste("Minimum:", min_val))
print(paste("Maximum:", max_val))

# Sum
sum_val <- sum(numeric_vector)
print(paste("Sum:", sum_val))

# Comprehensive Summary
print("\nSummary of numeric_vector:")
print(summary(numeric_vector))

# Correlation (requires two vectors)
vector_x <- c(1, 2, 3, 4, 5)
vector_y <- c(2, 4, 5, 4, 5)
cor_val <- cor(vector_x, vector_y)
print(paste("\nCorrelation between vector_x and vector_y:", cor_val))
```

### Data Visualization with `ggplot2`

`ggplot2` is one of the most popular and powerful packages for data visualization in R. It implements the 'Grammar of Graphics,' which provides a consistent and flexible way to create a wide variety of plots.

First, you need to load the `ggplot2` library. If you don't have it installed, you would run `install.packages("ggplot2")` (though it's often installed as part of the `tidyverse` package).

```r
library(ggplot2)
scatter_plot<-ggplot(data_frame_example,aes(x=Age,y=Name,color=IsStudent))+
  geom_point(size=4)+
  labs(title="Age  and Student Status by name",
  x="Age (Years)",
  y="Name")
  theme_minimal()
print(scatter_plot)
```

```r
bar_plot<-ggplot(data_frame_example,aes(x=IsStudent,fill=IsStudent))+
  geom_bar()+
  labs(title="Count of Students vs Non-Students",
  x="Is Student?",
  y="Count")+
    theme_bw()
print(bar_plot)
```

```r
set.seed(456)
data_for_hist_ggplot<-data.frame(Value=rnorm(100,mean=50,sd=10))
histogram_plot<-ggplot(data_for_hist_ggplot,aes(x=Value))+
  geom_histogram(binwidth=5,fill='darkgreen',color='black')+
  labs(title="Distribution of Random Numbers",
  x="Value",
  y="Frequency")+
    theme_classic()
print(histogram_plot)
```

### Reading Data from External Files

In real-world scenarios, your data will often come from external sources like CSV, Excel, or text files. R provides powerful functions to import this data. We'll focus on CSV files, as they are very common.

The `read.csv()` function (or `read.table()` with appropriate arguments) is primarily used for this purpose. Let's create a dummy CSV file first and then read it into an R data frame.

```r
# Create a data frame to write to CSV
sample_data_to_write <- data.frame(
  ID = c(1, 2, 3, 4, 5),
  Name = c("Alice", "Bob", "Charlie", "Diana", "Eve"),
  Score = c(85, 72, 90, 65, 78),
  Grade = c("A", "C", "A", "D", "B")
)

# Write the data frame to a CSV file
write.csv(sample_data_to_write, "sample_data.csv", row.names = FALSE)

print("Created 'sample_data.csv' with dummy data.")
```

```r
# Read the CSV file into an R data frame
sample_df <- read.csv("sample_data.csv")

# Print the data frame to see its contents
print("Data loaded from 'sample_data.csv':")
print(sample_df)

# Check the structure of the loaded data frame
print("\nStructure of the data frame:")
print(str(sample_df))
```
